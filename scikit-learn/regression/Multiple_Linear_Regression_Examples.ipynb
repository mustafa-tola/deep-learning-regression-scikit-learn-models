# **Exercise 1:-Restaurant Profit Prediction**

```
# Importing Libraries
import numpy as np
import pandas as pd
import sklearn
sklearn.__version__
```

```
# Reading Dataset
data = pd.read_csv("/content/Restaurant_Profit_Data.csv")
data.head()
```

```
# Check The Shape Of The Data
data.shape
```

```
# Create Feature Matrix And Dependent Variable Vector
X = data.iloc[:,:-1].values
y = data.iloc[:,4].values
```

```
# Display Value Of Feature Matrix
X
```

```
# Display Value Of Dependent Variabel Vector
y
```

```
# One Hot Encoding Of Categorical Column Called City
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
lblencoder = LabelEncoder()
X[:,3] = lblencoder.fit_transform(X[:,3]) 
```

```
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer


ct = ColumnTransformer(
    [('one_hot_encoder', OneHotEncoder(categories='auto'), [-1])],   # The column numbers to be transformed (here is [0] but can be [0, 1, 3])
    remainder='passthrough'                                         # Leave the rest of the columns untouched
)

X = ct.fit_transform(X)
```

```
# Display the feature matrix once the one hot encoding is done
X.astype(int)
```

```
# Get rid of the dummy variable (column called Chicago)
X = X[:,1:]
# This step is not in Linear regression model
```

```
# Display the feature matrix after the Chicago column
X.astype(int)
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```
len(X_train)
```

```
# Fitting Linear Regression Model To the Training set
from sklearn.linear_model import LinearRegression
MLR = LinearRegression()
MLR.fit(X_train,y_train)
```

```
# Print The Constant and Coefficients
print(f'intercept = {MLR.intercept_}')
print(f'constant = {MLR.coef_}')
```

```
# Predicting the test set results
y_predict = MLR.predict(X_test)
```

```
# Display actual and predicted values side by side
df = pd.DataFrame(data=y_test,columns=['actual'])
df['predicted'] = y_predict
df
```

```
# Predicting the single Observation Results.
singl_obs = [1,0,160349,134321,401400]
feature_array = np.array(singl_obs)
feature_array = feature_array.reshape(1,-1)
y_pred_single_obs = MLR.predict(feature_array)
round(float(y_pred_single_obs),2)
```

```
from sklearn.metrics import r2_score
r_square = r2_score(y_test,y_predict)
print(r_square)
```
